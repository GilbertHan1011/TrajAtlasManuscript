```python
df_new = adata_obs[["Sample","Age","Age(In Detail)","Organ", 'Tissue',
                    'Tissue(Specific)',"Stage","Bone Forming Methods","Origin",'Species']]
table=df_new.groupby("Sample")[['Age', 'Age(In Detail)', 'Organ', 'Tissue',
       'Tissue(Specific)', 'Stage', 'Bone Forming Methods', 'Origin',
       'Species']].first().reset_index()
table=table.set_index("Sample")
table=table.loc[df_wide.index.to_numpy(),: ]
```

```python
table["Age(In Detail)"] = pd.Categorical(table["Age(In Detail)"], categories=['E10', 'E11', 'E11.5', 'E12','8 WPC', 'E12.5', 'E13', '13 WPC', 
                                                                              '14 WPC', '15 WPC', '16 WPC','19 WPC', 
                                                                              'E13.5', 'E14', 'E14.5', 'E15', 'E15.5', 'E16', 'E16.5', 'E17', 'E17.5', 
                                                                              'E18', 'E18.5','P3', 'P4-P5',  'P5', 'P6', 'P7', 'P7.5','P10', 
                                                                              'P11', 'P13', '3W',  'P19', 
                                                                              'P21', 'P23', 'P25', 'P28','1M', 'P42', '1.5M','2-4M', '2M',
                                                                               '4W', '5 WPC', '6W', '7W',
                                                                              '8-12W', '8W',  "10W", '11-12W', '11W','10-12W', '10-14W',
                                                                              '9-12W', '12W','3M','4M','Adult',"12M",'12-14M',
                                                                               '16M', '18M'], ordered=True)
```

```python
table["Tissue(Specific)"] = pd.Categorical(table["Tissue(Specific)"])
table["Tissue"] = pd.Categorical(table["Tissue"])
table["Species"] = pd.Categorical(table["Species"])
table["Bone Forming Methods"] = pd.Categorical(table["Bone Forming Methods"])
```

```python
color_map_age = dict(zip([ 'Organogenesis stage','Fetal stage', 'Postnatal', 'Young Adult', 'Adult','Old'],
                         sns.color_palette("ch:s=.25,rot=-.25", n_colors=6)))
color_map_ageindetail = dict(zip(table["Age(In Detail)"].values.categories,
                         sns.color_palette("ch:s=.25,rot=-.25", n_colors=len(table["Age(In Detail)"].values.categories))))
color_map_organ = dict(zip(set(table["Organ"].values),
                         sns.color_palette('Spectral', n_colors=len(set(table["Organ"].values)))))
color_map_stage = dict(zip(table["Stage"].unique(),
                         sns.color_palette('Spectral', n_colors=len(table["Stage"].unique()))))
color_map_tissuespecific = dict(zip(table["Tissue(Specific)"].values.categories,
                         sns.color_palette('Spectral', n_colors=len(table["Tissue(Specific)"].unique()))))
color_map_tissue = dict(zip(table["Tissue"].values.categories,
                         sns.color_palette('Spectral', n_colors=len(table["Tissue"].values.categories))))
color_map_orgin = dict(zip(table["Origin"].unique(),
                         sns.color_palette('pastel', n_colors=len(table["Origin"].unique()))))
color_map_form = dict(zip(table["Bone Forming Methods"].values.categories,
                         sns.color_palette('pastel', n_colors=len(table["Bone Forming Methods"].values.categories))))
color_map_species = dict(zip(table["Species"].values.categories,
                         sns.color_palette('Set1', n_colors=len(table["Species"].values.categories))))
```

```python
from scipy.cluster.hierarchy import linkage, dendrogram
```

```python
linkage_matrix = linkage(df_wide, method='average')
dendrogram_result = dendrogram(linkage_matrix, labels=df_wide.index, orientation='top', distance_sort='descending', show_leaf_counts=True)
order = dendrogram_result['leaves']
```

```python
df_wide.iloc[order,[1,3,0,2]]
```

```python
import PyComplexHeatmap
from PyComplexHeatmap import *
annoRow=table
annoRow.index=df_wide.index
row_ha = HeatmapAnnotation(Age=anno_simple(annoRow["Age"],colors=color_map_age),
                           Age_detail=anno_simple(annoRow["Age(In Detail)"],colors=color_map_ageindetail),
                           Organ=anno_simple(annoRow["Organ"],colors=color_map_organ),
                           Tissue=anno_simple(annoRow["Tissue"],colors=color_map_tissue),
                           Tissue_specific=anno_simple(annoRow["Tissue(Specific)"],colors=color_map_tissuespecific),
                           Stage=anno_simple(annoRow["Stage"],colors=color_map_stage),
                           Origin=anno_simple(annoRow["Origin"],colors=color_map_orgin),
                           Bone_forming_methods=anno_simple(annoRow["Bone Forming Methods"],colors=color_map_form),
                           Species=anno_simple(annoRow["Species"],colors=color_map_species),axis=0)
plt.figure(figsize=(6, 8))

cm = ClusterMapPlotter(data=df_wide.iloc[order,[1,3,0,2]],right_annotation=row_ha,
                       linewidths=0.0001,cmap="GnBu",linecolor = 'white',
                       show_rownames=True,show_colnames=True,legend_width=50,col_cluster=False,row_cluster=False)
plt.savefig("../result/1.9_paga/paga_heatmap.pdf",dpi=1000, bbox_inches='tight')
```
